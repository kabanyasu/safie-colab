## 時間表記変更
- その都度修正する必要がある

```python
#出勤時間を秒数へ変換(数値検算済み)
import datetime
attendance_time_list = list(category_name_hour['出勤時間'])

second_list = []
#基準となる日時

#※その都度変更する必要がある
#シリアル値を日数へ変換すると、1900年1月1日を始点としている
#standard_date = datetime.datetime(1900,1,1)
standard_date = datetime.datetime(1899,12,31)

for time in attendance_time_list:
    try:
        #変換されている日時から、基準となる日時を差し引いて、求めたい時間のみを残す
        delta = time - standard_date
    except BaseException:
        second_list.append('NaN')
        continue
    second = delta.total_seconds()
    second = second/86400 #Excelは24時間=86400秒を1として計算しているから、こちらの処理でExcelで読み込めるようにする
    #閏年が関係している
    second_list.append(second)

# 表示したいときは以下をコメントアウトから戻す
#second_list
```

```python
category_name_hour['出勤時間'] = second_list
#category_name_hour
```

##PLをbiz,dev,pjの合計で割ってそれぞれに配布する

```python
import pandas as pd
PL = pd.read_excel(PL_excel,sheet_name = 'PL',header = None)
PL.columns = ['PL科目','PL科目金額']

PL
```

##合計値算出

```python
PJ = df_pj_sum['出勤時間'].sum()
BIZ = df_biz_sum['出勤時間'].sum()
DEV = df_dev_sum['出勤時間'].sum()

all_sum = PJ + BIZ + DEV
all_sum
```

```python
PL_value_list = list(PL['PL科目金額'])
PL_value_list

PL_pj_list = []
PL_biz_list = []
PL_dev_list = []

for PL_value in PL_value_list:
    divide_category_value = PL_value / all_sum
    
    #それぞれの値で掛けて、リストに入れる
    PL_pj = divide_category_value * PJ
    PL_biz = divide_category_value * BIZ
    PL_dev = divide_category_value * DEV

    PL_pj_list.append(PL_pj)
    PL_biz_list.append(PL_biz)
    PL_dev_list.append(PL_dev)
```

##それぞれのリストにある数値を一か所にまとめる

```python
df_PL_divide_category = pd.DataFrame([PL_pj_list, PL_biz_list, PL_dev_list]).T

df_PL_divide_category.columns = ['配賦用Biz','配賦用dev','配賦用PJ']

df_PL_divide_category
```

##PLシートの情報付与

```python
df_PL_divide_category_goal = pd.concat([PL,df_PL_divide_category],axis=1)
df_PL_divide_category_goal
```

```python
# 四捨五入して行う
df_PL_divide_category_goal['配賦用Biz'] = df_PL_divide_category_goal['配賦用Biz'].round()
df_PL_divide_category_goal['配賦用dev'] = df_PL_divide_category_goal['配賦用dev'].round()
df_PL_divide_category_goal['配賦用PJ'] =df_PL_divide_category_goal['PL科目金額'] - df_PL_divide_category_goal['配賦用Biz'] - df_PL_divide_category_goal['配賦用dev'] 
```

##biz配賦2段階目

```python
biz_unit1_sum =  df_PL_divide_category_goal['一単位当たり配賦用Biz'].sum()
biz_category_calc_list = []

for biz_name in df_biz_sum['出勤時間']:
    biz_category_sum = biz_name * biz_unit1_sum
    biz_category_calc_list.append(biz_category_sum)

df_biz_sum['配賦合計'] = biz_category_calc_list
```

```python
#bizをさらに配賦する
category_name_number = 0
#各PL科目の1単位分 * 各部門の出勤時間
for unit_1 in df_PL_divide_category_goal['一単位当たり配賦用Biz']:
    biz_unit_1_list = []
    #部門指定して出勤時間を掛ける
    for biz_time in df_biz_sum['出勤時間']:
        #1単位掛ける各部門
        calc_biz_unit_1 = unit_1 * biz_time
        biz_unit_1_list.append(calc_biz_unit_1)

    #bizのシート（部門名称、出勤時間、データ種別）の行方向に情報を足していく
    df_biz_sum[df_PL_divide_category_goal['PL科目'][category_name_number]] = biz_unit_1_list
    category_name_number = category_name_number+1

df_biz_sum = df_biz_sum.round()
df_biz_sum
```

##df_biz_sumを縦計して計算元と比較する。差額を任意の部門に足し引きする。

```python
#計算スペースを取得して、差を計算
df_biz_sum_calc_fraction = df_biz_sum.drop(['部門名称','出勤時間','データ種別','配賦合計'],axis=1)
calc_fraction = df_biz_sum.drop(['部門名称','出勤時間','データ種別','配賦合計'],axis=1).sum(axis=0)
calc_fraction = list(calc_fraction)
calc_fraction = df_PL_divide_category_goal['配賦用Biz'] - calc_fraction
#calc_fraction

#一番上の行で、足し引き算する
df_biz_change_fraction = list(df_biz_sum_calc_fraction.loc[0]) - calc_fraction
df_biz_sum_calc_fraction.loc[0] = list(df_biz_change_fraction)
df_biz_sum_calc_fraction
```

```python
#加工後（端数処理後）データと計算のため切り離した['部門名称','出勤時間','データ種別','配賦合計']を結合する
df_biz_sum = pd.concat([df_biz_sum[['部門名称','出勤時間','データ種別','配賦合計']],df_biz_sum_calc_fraction],axis=1)

df_biz_sum
```

```python
#計算スペースを取得して、差を計算
df_pj_sum_calc_fraction = df_pj_sum.drop(['部門名称','出勤時間','データ種別','配賦合計'],axis=1)
calc_fraction = df_pj_sum.drop(['部門名称','出勤時間','データ種別','配賦合計'],axis=1).sum(axis=0)
calc_fraction = list(calc_fraction)
calc_fraction = df_PL_divide_category_goal['配賦用PJ'] - calc_fraction
#calc_fraction

#一番上の行で、足し引き算する
df_pj_change_fraction = list(df_pj_sum_calc_fraction.loc[0]) - calc_fraction
df_pj_sum_calc_fraction.loc[0] = list(df_pj_change_fraction)
df_pj_sum_calc_fraction
```

```python
#加工後（端数処理後）データと計算のため切り離した['部門名称','出勤時間','データ種別','配賦合計']を結合する
df_pj_sum = pd.concat([df_pj_sum[['部門名称','出勤時間','データ種別','配賦合計']],df_pj_sum_calc_fraction],axis=1)

df_pj_sum
```
